## Collaborative Filtering Based Recommender System

```python
ratings_with_name
```

```python
pt = ratings_with_name.pivot_table(index='Event',columns='Skills',values='Scores')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
import pickle
pickle.dump(pt, open('pt.pkl', 'wb'))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
import pickle
pickle.dump(similarity_scores, open('similarity_scores.pkl', 'wb'))
```

```python
similarity_scores.shape
```

```python
def recommend(event_name):
    # index fetch
    index = np.where(pt.index==event_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:4]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = ratings_with_name[ratings_with_name['Event'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Event')['Event'].values))
        item.extend(list(temp_df.drop_duplicates('Event')['Date'].values))
        item.extend(list(temp_df.drop_duplicates('Event')['Location'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('Learn written English with us')
```

```python
recommend('Biology around us')
```

```python
recommend('Linear Algebra')
```

```python
recommend('Entrance Revision : NEET')
```
